```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('jsrt_metadata.csv', header=None)
```

```python
df
```

```python
X = df.iloc[:,9]
```

```python
size = len(X) - X.isnull().sum() -1
```

```python
X = df.iloc[1:size,9]
```

```python
X
```

```python
Y = X.unique()
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 100}


plt.rc('font', **font)

fig, ax = plt.subplots(figsize =(300, 120))
ax.hist(X, Y)

rects = ax.patches
labels = [Y[i] for i in range(len(rects))]
  
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height+0.01, label,
            ha='center', va='bottom')

plt.show()
```

```python
z = list()
for p in range(1,size):
    z.append(X[p])

dic = dict()
for word in z:
    if word in dic:
        dic[word] = dic[word]+1
    else:
        dic[word] = 1
dic
```

```python
font = {'family' : 'normal',
        'size'   : 10}


plt.rc('font', **font)

fig = plt.figure(figsize =(35, 18))
plt.pie(dic.values(), labels = dic.keys())
plt.legend(title = "Diseases: ")
plt.show()
```
